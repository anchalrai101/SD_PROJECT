```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nypd_data = pd.read_csv('NYPD_Arrests_Data__Historic__20241015.csv')
```

```python
#DESCRIBING DATA
print(nypd_data.describe())
```

```python
#INFORMATION OF DATASET
nypd_data.info()
```

```python
for col in nypd_data.select_dtypes(include='object'):
  mode_values = nypd_data[col].mode()

  if not mode_values.empty:
    nypd_data[col] = nypd_data[col].fillna(mode_values[0])
  else:
    nypd_data[col] = nypd_data[col].fillna('Unknown')
```

```python
for col in nypd_data.select_dtypes(exclude='object'):
  nypd_data[col] = nypd_data[col].fillna(nypd_data[col].mean())
```

```python
nypd_data.isnull().sum()
```

```python
import sqlite3

# Load data into SQLite for SQL queries
conn = sqlite3.connect(':memory:')
nypd_data.to_sql('nypd_arrests', conn, index=False, if_exists='replace')


query1 = """
SELECT * FROM nypd_arrests
WHERE OFNS_DESC = 'RAPE 1'
AND ARREST_DATE BETWEEN '2009-02-08' AND '2018-02-08'
"""
OFNS_DESC = pd.read_sql(query1, conn)
print("OFSN DESC:\n", OFNS_DESC.head())
```

```python
crime_counts = nypd_data.groupby('OFNS_DESC')['PD_CD'].count().reset_index()

# Rename columns for clarity
crime_counts.columns = ['Crime Type', 'Total Arrests']

# Sort by number of arrests (optional, for readability)
crime_counts = crime_counts.sort_values(by='Total Arrests', ascending=False)

# Display the result
print("Total Arrests by Crime Type:\n", crime_counts)
```
